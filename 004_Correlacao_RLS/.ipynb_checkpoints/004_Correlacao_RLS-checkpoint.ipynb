# Criar a regressão linear

## Defina quais serão suas variaveis dependentes e as independentes

```python
list(dados)
```

Separar dados que não foram afetados por disturbios

```python
nonPAD = dados[dados["Disturbance"] == 0]
```

Observe o conjunto de dados

```python
nonPAD.head()
```

Selecione os dados para análise

```python
y = nonPAD ['LeapRate'] # Taxa de saltos (numero de saltos/minuto)
x1= nonPAD ['DawnTime'] # Horário relativo da gravação (quantas horas se passaram desde o alvorecer)
```

Vai padronizar o conjunto de dados? Resposta = Sim

Criar função para padronizar conjunto de dados

```python
def escala(ConjundoDados):
    meanX = np.average(ConjundoDados) # Calcule a media de ConjundoDados
    sdX= np.std(ConjundoDados) # Calcule sd de ConjundoDados
    ConjundoDados.index =  list(range(len(ConjundoDados)))
    aux = {}
    for i in range(len(ConjundoDados)):
        aux[i] =(ConjundoDados[i]-meanX)/sdX
    reposta = list(aux.values())
    return reposta 
```

```python
y = escala(y)
x1 = escala(x1)
```

```python
Stand_nonPAD=pd.DataFrame({"y":y,"x1":x1})
```

```python
Stand_nonPAD.head()
```

Caso você queira exportar esses dados para algum propósito, você pdoe criar data frame com os dados padronizados. Em seguir faz a exportação. OBS: Nesse tutorial não precisamos exportar esses dados, mas vou mostrar como faz para fins didáticos.

```python
Stand_nonPAD.to_csv ('Dados_padronizados.csv', index = False, header=True)
```

## Explorar dados

Gráfico de dispersão (scatter plot). OBS: Dados padronizados

```python
plt.scatter(x1,y) # Fazer o plot
plt.xlabel('Standardized time since dawn', fontsize = 20) # Renomear eixo x
plt.ylabel('Standardized leap rate', fontsize = 20) # Renomear eixo y
plt.show() # Mostrar o plot
```

## Regressão

Adicione uma constante, para que seu modelo contenha a estimativa do parametro B0 (ponto de intersecção em y, quando x=0) 

```python
x = sm.add_constant(Stand_nonPAD["x1"])
```

```python
x.head()
```

Veja como são os valores de cada displar

Faça o modelo de regressão linar (OLS: ordinary least squares), o qual y será a variável dependente e x será a variável independente

```python
resultado = sm.OLS(y,x).fit()
```

Veja um resumo do resultado da sua regressão

```python
resultado.summary()
```

Vamos fazer uma grafico para mostrar a relação

```python
plt.scatter(x1,y) # Fazer o plot
yhat = -0.4348*x["x1"] # reta de regressão 
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line') # Inserir a reta de regressão no gráfico
plt.xlabel('Standardized time since dawn', fontsize = 20) # Renomear eixo x
plt.ylabel('Standardized leap rate', fontsize = 20) # Renomear eixo y
plt.show() # Mostrar o plot
```
